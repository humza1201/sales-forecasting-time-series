```python
res_df = pd.DataFrame({
    'date': y_test.index,
    'actual': y_test.values,
    'predicted': predictions,
    'residual': residuals
}).set_index('date')
```

```python
res_df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(res_df.index, res_df['residual'])
plt.xlabel('date')
plt.ylabel('residuals')
plt.axhline(0, linestyle='--')
plt.title('residuals over time')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(res_df['residual'], bins=30, edgecolor='black')
plt.title("Distribution of Residuals")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(res_df.index, res_df['actual'], label='Actual')
plt.plot(res_df.index, res_df['predicted'], label='Predicted')
plt.legend()
plt.title("Actual vs Predicted Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(res_df['predicted'], res_df['residual'], alpha=0.6)
plt.axhline(0, linestyle='--')
plt.xlabel("Predicted Sales")
plt.ylabel("Residual")
plt.title("Residuals vs Predicted")
plt.show()
```
